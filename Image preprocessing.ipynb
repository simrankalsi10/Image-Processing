```python
def list_min_max(int_list):

    our_list = []

    max_int = max(int_list)
    min_int = min(int_list)
    pct_10, pct_15, pct_25 = [pct* (max_int-min_int) for pct in [0.1, 0.15, 0.25]]

    first = [min_int, min_int+pct_10]
    second = [first[1], first[1]+pct_15]
    third = [second[1], second[1]+pct_25]
    fourth = [third[1], third[1]+pct_25]
    fifth = [fourth [1],max_int]

    output = [first, second, third, fourth, fifth]

#     print (output)
    return output
```
